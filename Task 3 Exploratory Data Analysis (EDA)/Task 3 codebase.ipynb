### Step 1: Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (Titanic dataset used for demonstration)
df = pd.read_csv(r"C:\Users\hp\Codveda Projects\Task 3 Exploratory Data Analysis (EDA)\Data\cleaned_train.csv")

# Preview dataset
df.head()
```

### Step 2: Summary Statistics

```python
# General statistics
print(df.describe(include="all"))

# Specific statistics
print("Mean Age:", df["Age"].mean())
print("Median Age:", df["Age"].median())
print("Fare Variance:", df["Fare"].var())
```

### Step 3: Univariate Analysis (Histograms & Boxplots)

```python
# Age distribution
plt.figure(figsize=(6,4))
sns.histplot(df["Age"], bins=20, kde=True)
plt.title("Age Distribution")
plt.savefig(r"C:\Users\hp\Codveda Projects\Task 3 Exploratory Data Analysis (EDA)\Visuals\age_distribution.png", dpi=300, bbox_inches='tight')
plt.show()

# Fare distribution
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Fare"])
plt.title("Fare Distribution")
plt.savefig(r"C:\Users\hp\Codveda Projects\Task 3 Exploratory Data Analysis (EDA)\Visuals\Fare Distribution.png", dpi=300, bbox_inches='tight')
plt.show()
```

### Step 4: Bivariate Analysis (Scatter Plots & Boxplots)

```python
# Age vs Fare
plt.figure(figsize=(6,4))
sns.scatterplot(x="Age", y="Fare", data=df, hue="Survived")
plt.title("Age vs Fare (colored by Survival)")
plt.savefig(r"C:\Users\hp\Codveda Projects\Task 3 Exploratory Data Analysis (EDA)\Visuals\Age vs Fare.png", dpi=300, bbox_inches='tight')
plt.show()

# Survival by Passenger Class
plt.figure(figsize=(6,4))
sns.boxplot(x="Pclass", y="Fare", data=df)
plt.title("Fare Distribution by Passenger Class")
plt.savefig(r"C:\Users\hp\Codveda Projects\Task 3 Exploratory Data Analysis (EDA)\Visuals\Fare Distribution by Passenger Class.png", dpi=300, bbox_inches='tight')
plt.show()
```

### Step 5: Correlation Analysis

```python
# Correlation matrix
corr = df.corr(numeric_only=True)

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.savefig(r"C:\Users\hp\Codveda Projects\Task 3 Exploratory Data Analysis (EDA)\Visuals\Correlation Matrix.png", dpi=300, bbox_inches='tight')
plt.show()
```
